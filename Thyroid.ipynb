```python
def iqr_capping(df,cols,factor):

    for col in cols:

        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)

        iqr = q3 - q1

        upper_bound = q3 + (factor*iqr)
        lower_bound = q1 - (factor*iqr)

        df[col] = np.where(df[col]>upper_bound,upper_bound,
                 np.where(df[col]<lower_bound,lower_bound,df[col]))
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np

y_probas = svm_model.predict_proba(X_test)

# Calculate and plot the Cumulative Gain Curve for each class
def cumulative_gain(y_true, y_probas, positive_class):
    sorted_indices = np.argsort(y_probas[:, positive_class])[::-1]
    y_true_sorted = y_true[sorted_indices]
    gains = np.cumsum(y_true_sorted) / np.sum(y_true_sorted)

    plt.plot(np.linspace(0, 1, len(gains)), gains, label=f'Class {positive_class}')

plt.figure(figsize=(8, 6))

# Calculate and plot cumulative gains for each class
for i in range(len(np.unique(Y_test))):
    cumulative_gain((Y_test == i).astype(int), y_probas, i)

plt.xlabel('Percentage of samples (sorted by probability)')
plt.ylabel('Cumulative Gain')
plt.title('Cumulative Gain Curve for Multi-class Classification')
plt.legend()
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np

y_probas = svm_model.predict_proba(X_test)

# Calculate and plot the Lift Curve for each class
def lift_curve(y_true, y_probas, positive_class):
    sorted_indices = np.argsort(y_probas[:, positive_class])[::-1]
    y_true_sorted = y_true[sorted_indices]
    total_positives = np.sum(y_true)
    gains = np.cumsum(y_true_sorted) / total_positives

    # Calculate the ideal case where all positives are at the top
    ideal_positives = np.arange(1, len(y_true) + 1)
    ideal_gains = np.cumsum(np.sort(y_true)[::-1]) / total_positives
    ideal_lift = gains / ideal_gains

    plt.plot(np.linspace(0, 1, len(gains)), ideal_lift, label=f'Class {positive_class}')

plt.figure(figsize=(8, 6))

# Calculate and plot Lift curve for each class
for i in range(len(np.unique(Y_test))):
    lift_curve((Y_test == i).astype(int), y_probas, i)

plt.xlabel('Percentage of samples (sorted by probability)')
plt.ylabel('Lift')
plt.title('Lift Curve for Multi-class Classification')
plt.legend()
plt.show()
```
